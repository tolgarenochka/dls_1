```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
#посмотрим на категориальные признаки
data[cat_cols].nunique()
```

```python
# поработаем с признаком пола. заменим строковые значения муж/жен на 0/1
data['Sex_encoded'] = data['Sex'].apply(lambda x: 1 if x == "Female" else 0)

data.drop(columns=['Sex'], inplace=True)
```

```python
# поработаем с признаком наличия партнера. заменим строковые значения да/нет на 1/0
data['HasPartner_encoded'] = data['HasPartner'].apply(lambda x: 1 if x == "Yes" else 0)

data.drop(columns=['HasPartner'], inplace=True)
```

```python
# поработаем с признаком наличия бейбика. заменим строковые значения да/нет на 1/0
data['HasChild_encoded'] = data['HasChild'].apply(lambda x: 1 if x == "Yes" else 0)

data.drop(columns=['HasChild'], inplace=True)
```

```python
# поработаем с признаком HasPhoneService. заменим строковые значения да/нет на 1/0
data['HasPhoneService_encoded'] = data['HasPhoneService'].apply(lambda x: 1 if x == "Yes" else 0)

data.drop(columns=['HasPhoneService'], inplace=True)
```

```python
# поработаем с признаком IsBillingPaperless. заменим строковые значения да/нет на 1/0
data['IsBillingPaperless_encoded'] = data['IsBillingPaperless'].apply(lambda x: 1 if x == "Yes" else 0)

data.drop(columns=['IsBillingPaperless'], inplace=True)
```

```python
# поработаем с оставшимися не бинарными полями. для этого воспользуемся one hot encoding
one_hot_columns = ['HasInternetService', 'HasMultiplePhoneNumbers', 'HasOnlineSecurityService',
                   'HasOnlineBackup', 'HasDeviceProtection', 'HasTechSupportAccess',
                   'HasOnlineTV', 'HasMovieSubscription', 'HasContractPhone', 'PaymentMethod']
categorical_features = data[one_hot_columns]
unprocessed_cat_features = categorical_features.select_dtypes(include=[object]).columns.tolist()

#Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Apply one-hot encoding to the categorical columns
one_hot_encoded = encoder.fit_transform(data[unprocessed_cat_features])

#Create a DataFrame with the one-hot encoded columns
#We use get_feature_names_out() to get the column names for the encoded data
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(unprocessed_cat_features), index=data.index)

data = pd.concat([data, one_hot_df], axis=1)
data = data.drop(unprocessed_cat_features, axis=1)
```

```python
data.info()
```

```python
# Обратим внимание на столбцы с окончанием _No internet service. Есть подозрение, что они идентичны
# Проверяем, есть ли полностью одинаковые столбцы
is_equal_columns = ['HasOnlineSecurityService_No internet service', 'HasOnlineBackup_No internet service', 'HasDeviceProtection_No internet service',
           'HasTechSupportAccess_No internet service', 'HasOnlineTV_No internet service', 'HasMovieSubscription_No internet service']
for i in range(1, len(is_equal_columns)):
    is_equal = (data[is_equal_columns[0]] == data[is_equal_columns[i]]).all()
    print(f"{is_equal_columns[0]} == {is_equal_columns[i]}: {is_equal}")
```

```python
# Тогда создаим общий столбец, а дубли удалим

data['NoInternetService'] = data['HasOnlineSecurityService_No internet service']

data.drop(columns=is_equal_columns, inplace=True)
```

```python
X = data.drop(columns=['Churn'])
y = data['Churn']
```

```python
# Создаем pipeline с масштабированием и логистической регрессией
pipeline = make_pipeline(
    StandardScaler(),  # Масштабируем данные
    LogisticRegression(random_state=42, max_iter=1000)
)

# Параметры для поиска
param_grid = {
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001]
}

# Создаем GridSearchCV с кроссвалидацией
grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,  # 5-фолдовая кроссвалидация
    scoring='roc_auc',
    n_jobs=-1,  # используем все ядра процессора
    return_train_score=True
)

# Запускаем поиск лучшего параметра C
print("Запуск кроссвалидации для подбора параметра C...")
grid_search.fit(X, y)

# Выводим результаты
print("\n" + "="*50)
print("РЕЗУЛЬТАТЫ КРОССВАЛИДАЦИИ")
print("="*50)

print(f"Лучший параметр C: {grid_search.best_params_['logisticregression__C']}")
print(f"Лучшее ROC-AUC: {grid_search.best_score_:.4f}")

# Детальные результаты по всем C
results_df = pd.DataFrame(grid_search.cv_results_)
print("\nДетальные результаты:")
for i, row in results_df.iterrows():
    c_value = row['param_logisticregression__C']
    mean_test = row['mean_test_score']
    std_test = row['std_test_score']
    print(f"C = {c_value:6}: ROC-AUC = {mean_test:.4f} ± {std_test:.4f}")
```

**Лучшее качество - 0.8451 при С = 100**

```python
# 2. Подбор гиперпараметров
print("\n2. Подбор гиперпараметров:")

# Параметры для поиска
n_estimators_list = [100, 200, 500]
learning_rate_list = [0.01, 0.05, 0.1]

best_auc = 0
best_params = {}
best_model = ''

for n_estimators in n_estimators_list:
    for learning_rate in learning_rate_list:
        print(f"Тестируем: n_estimators={n_estimators}, learning_rate={learning_rate}")

        model = CatBoostClassifier(
            n_estimators=n_estimators,
            learning_rate=learning_rate,
            cat_features=cat_cols,
            random_state=42,
            verbose=0
        )

        model.fit(X_train, y_train)
        y_pred_proba = model.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_val, y_pred_proba)

        print(f"  ROC-AUC: {auc:.4f}")

        if auc > best_auc:
            best_auc = auc
            best_params = {'n_estimators': n_estimators, 'learning_rate': learning_rate}
            best_model = model

print("\n" + "="*50)
print("РЕЗУЛЬТАТЫ ПОДБОРА:")
print("="*50)
print(f"Лучшая комбинация параметров:")
print(f"  n_estimators: {best_params['n_estimators']}")
print(f"  learning_rate: {best_params['learning_rate']}")
print(f"  Лучший ROC-AUC: {best_auc:.4f}")
```

Лучшая комбинация параметров:
  n_estimators: 200,
  learning_rate: 0.05,
  Лучший ROC-AUC: 0.8565
